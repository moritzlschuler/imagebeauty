```
import glob
import shutil
from sklearn.model_selection import train_test_split
```

```
DATA_DIR = "./drive/My Drive/imagebeauty/data/GPA/"
```

```
pos_img_paths = [p for p in glob.glob(DATA_DIR + 'Pos_Resize' + '/*.jpg')]
neg_img_paths = [p for p in glob.glob(DATA_DIR + 'Neg_Resize' + '/*.jpg')]
```

```
pos_train, pos_test = train_test_split(pos_img_paths, train_size = 0.7)
neg_train, neg_test = train_test_split(neg_img_paths, train_size = 0.7)
print(pos_train[:10])
print(len(pos_train))
print(len(pos_test))
```

```
for i in pos_train:
  shutil.move(i, './drive/My Drive/imagebeauty/data/GPA/training/positive/')
```

```
for i in neg_train:
  shutil.move(i, './drive/My Drive/imagebeauty/data/GPA/training/negative/')
```

```
for i in pos_test:
  shutil.move(i, './drive/My Drive/imagebeauty/data/GPA/testing/positive/')
```

```
for i in neg_test:
  shutil.move(i, './drive/My Drive/imagebeauty/data/GPA/testing/negative/')
```
